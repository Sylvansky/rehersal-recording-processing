# Phase 2: Stem Separation (Demucs)
Separate recordings into individual stems (vocals, drums, bass, other) using GPU-accelerated Demucs.

```python
from pipeline.config import load_config
from pipeline.stem import find_audio_files, run_stem_separation, list_stems
```

```python
config = load_config()
print(f"Device:      {config['DEVICE']}")
print(f"Batch size:  {config['BATCH_SIZE']}")
print(f"Num workers: {config['NUM_WORKERS']}")
```

## Find Audio Files

```python
audio_files = find_audio_files(config["RAW_DIR"])
print(f"Found {len(audio_files)} audio files")
for f in audio_files[:10]:
    print(f"  {f}")
```

## Run Stem Separation
⚠️ This is the most time-consuming step. With 110h of audio on RTX 5070 Ti, expect ~20-30 hours of GPU processing.

```python
run_stem_separation(
    input_dir=config["RAW_DIR"],
    output_dir=config["STEMS_DIR"],
    device=config["DEVICE"],
    batch_size=config["BATCH_SIZE"],
    num_workers=config["NUM_WORKERS"],
)
```

## Review Results

```python
stems = list_stems(config["STEMS_DIR"])
print(f"Total stems: {len(stems)}")
```
